### Detecting and removing outliers in approved_budget variable
1. Method used: IQR (Interquartile Range)
2. Calculating Q1, Q3 (i.e. 25th and 75th percentile)
3. Calculating IQR as Q3 - Q1
4. Defining bounds for outliers as Q1 - 1.5 * IQR and Q3 + 1.5 * IQR
5. Creating a new dataframe without outliers
6. Printing the number of outliers removed
7. Printing summary statistics before and after outlier removal
8. Visualizing the box plot before and after outlier removal


```python


Q1 = df['approved_budget'].quantile(0.25)
Q3 = df['approved_budget'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new dataframe without outliers
df_no_outliers = df[(df['approved_budget'] >= lower_bound) & (df['approved_budget'] <= upper_bound)]

# Print the number of outliers removed
print(f"Number of outliers removed: {len(df) - len(df_no_outliers)}")

# Print summary statistics before and after outlier removal
print("\nBefore outlier removal:")
print(df['approved_budget'].describe())

print("\nAfter outlier removal:")
print(df_no_outliers['approved_budget'].describe())

# Optionally, create a new box plot without outliers
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.boxplot(df_no_outliers['approved_budget'], vert=False)
plt.title('Box Plot of Approved Budget (Without Outliers)')
plt.xlabel('Approved Budget')
plt.ylabel('Variable')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
